```python
path="data/"
for folder in os.listdir(path):
    #print(folder)
    if os.path.isdir(os.path.join(path,folder)):
        print(folder)
        i=0
        for file in os.listdir(os.path.join(path,folder)):
            loc=os.path.join(path,folder,file)
            print(loc)
            print(os.path.join(path,folder)+"/"+str(i)+".jpg")
            os.rename(loc,os.path.join(path,folder)+"/"+str(i)+".jpg")
            i+=1
        print(i)
```
